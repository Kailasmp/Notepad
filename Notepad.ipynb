```python
import tkinter as tk
from tkinter import Text, END
from tkinter.messagebox import showinfo
from tkinter.filedialog import askopenfilename, asksaveasfilename
import os

# Function definitions
def new_file():
    global file
    root.title("Untitled - Notepad")
    file = None
    T.delete(1.0, END)

def open_file():
    global file
    file = askopenfilename(defaultextension=".txt", filetypes=[("All Files", "*.*"), ("Text Documents", "*.txt")])
    if file:
        root.title(os.path.basename(file) + " - Notepad")
        T.delete(1.0, END)
        with open(file, "r") as f:
            T.insert(1.0, f.read())

def save_file():
    global file
    if file is None:
        file = asksaveasfilename(initialfile="Untitled.txt", defaultextension=".txt", filetypes=[("All Files", "*.*"), ("Text Documents", "*.txt")])
        if file:
            with open(file, "w") as f:
                f.write(T.get(1.0, END))
            root.title(os.path.basename(file) + " - Notepad")
            showinfo("Save", "File Saved Successfully")
    else:
        with open(file, "w") as f:
            f.write(T.get(1.0, END))
        showinfo("Save", "File Saved Successfully")

def cut():
    T.event_generate("<<Cut>>")

def copy():
    T.event_generate("<<Copy>>")

def paste():
    T.event_generate("<<Paste>>")

def about():
    showinfo("Notepad", "Notepad Made By Kailas")

def exit_app():
    root.destroy()

# Initialize main application window
root = tk.Tk()
root.title("Untitled - Notepad")
try:
    root.wm_iconbitmap("notepad.ico")  # Ensure this is a .ico file
except Exception:
    print("Icon file not found")

root.geometry("500x500")

# Create a Text widget
T = Text(root, font="lucida")
T.pack(expand=True, fill=tk.BOTH)
file = None

# Create menu bar
Menubar = tk.Menu(root)

# Start of file menu
fileMenu = tk.Menu(Menubar, tearoff=0)
fileMenu.add_command(label="New", command=new_file)
fileMenu.add_command(label="Open", command=open_file)
fileMenu.add_command(label="Save", command=save_file)
fileMenu.add_separator()
fileMenu.add_command(label="Quit", command=exit_app)
Menubar.add_cascade(label="File", menu=fileMenu)

# Start of edit menu
editMenu = tk.Menu(Menubar, tearoff=0)
editMenu.add_command(label="Cut", command=cut)
editMenu.add_command(label="Copy", command=copy)
editMenu.add_command(label="Paste", command=paste)
Menubar.add_cascade(label="Edit", menu=editMenu)

# Start of help menu
helpMenu = tk.Menu(Menubar, tearoff=0)
helpMenu.add_command(label="About Notepad", command=about)
Menubar.add_cascade(label="Help", menu=helpMenu)

# Configure the menu bar
root.config(menu=Menubar)

# Run the main loop
root.mainloop()
```
